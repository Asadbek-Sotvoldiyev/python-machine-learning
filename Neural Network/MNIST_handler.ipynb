```python
import pandas as pd

train = pd.read_csv("datasets/mnist/mnist_train.csv")
test = pd.read_csv("datasets/mnist/mnist_test.csv")
```

```python
from collections import Counter

train_count = Counter(train['label'])
train_count
```

### Train dataset visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="label", data=train)
plt.title("Distribution of classes in MNIST train dataset")
plt.xlabel("Classes")
plt.ylabel("Number of samples")
plt.show()
```

### Test dataset vizualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="label", data=test)
plt.title("Distribution of classes in MNIST test dataset")
plt.xlabel("Classes")
plt.ylabel("Number of samples")
plt.show()
```
